# **1. While Loop**

## **1.1 Mô tả**

- Sự ngẫu nhiên là một đặc điểm cơ bản của các hiện tượng trong thực tế. Nó thể
hiện qua việc kết quả của một thí nghiệm không thể dự đoán trước được một
cách chính xác. VD:
  - Chọn ngẫu nhiên hai số a và b sao cho tổng của $a + b = 40$, câu hỏi đặt ra rằng, chúng ta phải tạo ngẫu nhiên bao nhiêu lần để thỏa mản điều kiện trên ?

- **Cấu trúc vòng lặp While-Loop**: Vòng lặp while là một cấu trúc điều khiển
trong Python cho phép thực thi một khối mã nhiều lần miễn là điều kiện cho
trước vẫn còn đúng.

## **1.2 Bài tập: về sự ngẫu nhiên - Dùng vòng lặp While**

- Chọn ngẫu nhiên hai số a và b thuộc từ 1 đến 20 sao cho tổng của $a + b = 40$, câu hỏi đặt ra rằng, chúng ta phải tạo ngẫu nhiên bao nhiêu lần để thỏa mản
điều kiện trên.

```python
import random
```

```python
def random_number_with_condition(target):
  # Đặt seed để kết quả ngẫu nhiên giống nhau ở mọi máy
  random.seed(0)

  # Khởi tạo biến đếm số lần thử
  count = 0

  # Vòng lặp while vô hạn, chỉ dừng khi thoả mãn điều kiện
  while True:
    # Mỗi lần lặp là một lần tạo số mới
    count += 1

    # Chọn ngẫu nhiên a và b từ 1 đến 20
    a = random.randint(1, 20)
    b = random.randint(1, 20)

    if a + b == target:
      return(f"[target={target}] Tìm thấy: a={a}, b={b} → {count} lần")
```

```python
print(f"Test case 1: {random_number_with_condition(40)}")
print(f"Test case 2: {random_number_with_condition(20)}")
print(f"Test case 3: {random_number_with_condition(35)}")
```

# **2. While Loop + continue and break keywords**

## **2.1 Mô tả**

- Trong Python, câu lệnh **while** được sử dụng để tạo một vòng lặp, trong đó các biểu thức được kiểm tra và nếu điều kiện đúng, khối mã lệnh bên trong vòng lặp sẽ được thực thi. Các lệnh **break** và **continue** được sử dụng để kiểm soát luồng của vòng lặp.
  - Câu lệnh **break** được sử dụng để thoát khỏi vòng lặp ngay lập tức.
  - Câu lệnh **continue** được sử dụng để bỏ qua phần còn lại của vòng lặp và chuyển
đến lần lặp tiếp theo.

## **2.2 Bài tập**

- Cài đặt hàm find_divisible_number(a) tìm số nguyên dương nhỏ nhất lớn hơn 100 và chia hết cho số nguyên dương a

```python
def find_divisible_number(a):
  # tìm số nguyên dương nhỏ nhất lớn hơn 100 -> nên chúng ta sẽ khởi tạo 101
  smallest_int = 101

  # Lặp vô hạn cho đến khi tìm thấy số thỏa mãn
  while True:

    # Kiểm tra xem số hiện tại có chia hết cho `a` hay không
    if smallest_int % a == 0:
      # Nếu chia hết thì trả về kết quả ngay lập tức và thoát khỏi function
      return smallest_int

    # Tăng lên 1 để kiểm tra số tiếp theo
    smallest_int += 1
```

```python
print(f"Test case 1: {find_divisible_number(5)}")
print(f"Test case 2: {find_divisible_number(17)}")
```

## **3. Phương pháp Newton (Newton’s Method)**

## **3.1 Mô tả**

- Phương pháp Newton (Newton’s Method), còn được gọi là phương pháp Newton-
Raphson, là một phương pháp số học để tìm gần đúng của các nghiệm của một hàm số thực:
  - Cụ thể, nó thường được sử dụng để tìm gần đúng của các nghiệm
của phương trình $f(x) = 0$.

- Ngoài ứng dụng trong tìm nghiệm của một hàm số, phương pháp Newton còn có
ứng dụng trong máy học (Machine learning) trong việc tìm nghiệm của đạo hàm
của hàm loss. Tuy nhiên đây là phương pháp không phổ biến bằng thuật toán
gradient descent.

- Chúng ta sẽ dùng phương pháp Newton để tính căn bậc hai cho một
số dương a. Chúng ta thực hiện các bước sau:

  - Khởi tạo giá trị $x_0 = a, n = 0$ và cho trước sai số $\varepsilon$ (thực ra $x_{0}$ có thể nhận bất kỳ giá trị dương nào). Tiếp đó, ta sẽ đi xây dựng hàm $f(x) = x^2 - a$. Ở đây, ta xem $x_{n}$ (ở bước hiện tại đang là $x_{0}$ chính là lời giải cho bài toán tính căn bậc hai của $a$.

  - Cải thiện xấp xỉ $x_{n}$ bằng xấp xỉ $x_{n+1}$ theo công thức tổng quát như sau: $x_{n+1}=x_{n}-\frac{f(x_{n})}{f^{\prime }(x_{n})}$

  - So sánh xấp xỉ $x_{n+1}$ và $x_{n}$. Nếu $\vert{}x_{n+1} - x_n\vert{} < \varepsilon$, ta sẽ dừng việc cải thiện xấp xỉ, và trả về kết quả căn bậc hai của $a$ là $x_{n+1}$. Ngược lại, thực hiện tiếp bước (b) với $n = n + 1$.

## **3.2 Bài tập**

-  Cài đặt hàm find_squared_root(a) tìm căn bậc hai cho một số a bất kì với
$ε = 0.001$.

```python
def find_squared_root(a):
  """ Find the squared root of number a """
  if a < 0:
    return "Không tìm được căn bậc hai của số âm"

  if a == 0:
    return 0

  EPSILON = 0.001

  x_0 = a

  while True:
    f = x_0 ** 2 - a  # f(x)  = x² - a
    dfx = 2 * x_0     # f'(x) = 2x

    x_new = x_0 - f / dfx

    if abs(x_new - x_0) < EPSILON: # |xn+1 − xn| < ε,
      return x_new

    x_0 = x_new  # n = n + 1.
```

```python
print(f"Test case 1: {find_squared_root(2)}")
print(f"Test case 2: {find_squared_root(3)}")
print(f"Test case 3: {find_squared_root(-2)}")
print(f"Test case 4: {find_squared_root(0)}")
```
