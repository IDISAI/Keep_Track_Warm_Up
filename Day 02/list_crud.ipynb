Câu 1: Tạo mới một List có tên là lst_data, gồm các số chẵn từ 1 đến 12.

```python
lst_data = list(range(2,12,2))
print(lst_data)
```

Câu 2: Xóa tất cả các số chia hết cho 3 trong lst_data vừa tạo

```python
lst_data = [x for x in lst_data if x % 3 != 0]
print(lst_data)
```

Câu 3: Thêm vào cuối lst_data các số từ 1 đến 3, và thêm vào vị trí index = 3
một chuỗi các số từ 6 đến 8

```python
# Thêm vào cuối lst_data các số từ 1 đến 3,
lst_data.extend(range(1, 4))

# thêm vào vị trí index = 3 một chuỗi các số từ 6 đến 8
lst_data[3:3] = range(6, 9)
print(lst_data)
```

Câu 4: Nếu các số trong list lst_data chia hết cho 2 hoặc chia hết cho 5 thì cập
nhật thành số 0

```python
update_lst_data_1 = [0 if x % 2 == 0 or x % 5 == 0 else x for x in lst_data]
print(f"List Comprehension: {update_lst_data_1}")
```
