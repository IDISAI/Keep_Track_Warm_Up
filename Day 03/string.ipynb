- Câu 1:Tạo một biến name gán giá trị là tên một người, sau đó hiển thị ra màn hình một thông báo chứa tên đó sử dụng `f-string`, VD:

````python
name = "Alice"
print(f"{name} is a great teacher!")
````

- Câu 2: Tạo một biến và gán giá trị `"ms Taylor"` cho nó, sau đó thực hiện in ra màn hình giá trị biến đó được viết hoa ký tự đầu tiên mỗi từ hoặc viết toàn bộ bằng chữ chữ hoa, hoặc chữ thường.

```pthon

Input: name = “ms Taylor”
Output:
  Title case: Ms Taylor
  Upper case: MS TAYLOR
  Lower case: ms taylor
```

````python
name = "ms Taylor"
print(f"Title case: {name.title()}")
print(f"Upper case: {name.upper()}")
print(f"Lower case: {name.lower()}")
````
