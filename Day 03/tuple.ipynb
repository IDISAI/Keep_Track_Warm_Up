## **1. Khởi tạo Tuple và thao tác tìm kiếm, trích xuất thông tin trên Tuple đó.**

- Câu 1: Tạo mới hai Tuple: `my_tuple1 = (2,3)`, `my_tuple2 = (3,6)` mỗi Tuple có 2 phần tử đại diện cho một vector trong không gian 2D.

```python
my_tuple1 = (2,3)
my_tuple2 = (3,6)
```

- Câu 2: In ra kết quả của tổng và tích 2 vector trên.

```python
def tupe_sum(a,b):
  return tuple(x + y for x, y in zip(a,b))

def tupe_product(a,b):
  return tuple(x * y for x, y in zip(a,b))

print(tupe_sum(my_tuple1,my_tuple2))
print(tupe_product(my_tuple1,my_tuple2))
```

- Câu 3: In ra kết quả của khoảng cách của hai vector trên theo công thức. Biết:
$$\mathrm{distance}(P, Q)=\sqrt{(p_1-q_1)^2+(p_2-q_2)^2+\cdots+(p_n-q_n)^2}$$

```python
import math

# distance = math.sqrt((my_tuple1[0]-my_tuple2[0])**2 + (my_tuple1[1]-my_tuple2[1])**2)
# print(f"Khoảng cách: {distance}")

distance  = lambda a, b: math.sqrt(sum([(q1-q2)**2 for q1, q2 in zip(a, b)]))
print(f"Khoảng cách: {distance(my_tuple1, my_tuple2)}")
```

- Câu 4: In ra vị trí của phần tử có giá trị là 3:

  - Sử dụng cú pháp: `my_tuple.index(values)` để trích xuất vị trí của giá trị cần tìm.

```python
idx1 = my_tuple1.index(3)  # vị trí 1
idx2 = my_tuple2.index(3)  # vị trí 0
print(f"Vị trí của 3 trong my_tuple1: {idx1}, trong my_tuple2: {idx2}")
```
