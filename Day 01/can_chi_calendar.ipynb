- Lịch Can Chi là một hệ thống tính toán giờ, ngày, tháng năm âm lịch của người Trung Quốc cổ đại. Can chi có 10 thiên can và 12 địa chi:
  - 10 Thiên can: Canh, Tân, Nhâm, Quý, Giáp, Ất, Bính, Đinh, Mậu, Kỷ.
  - 12 địa chi: Thân, Dậu, Tuất, Hợi, Tý, Sửu, Dần, Mão, Thìn, Tỵ, Ngọ, Mùi.

# **Bài tập: Tính lịch Can Chi - làm quen với câu lệnh `if/else` hoặc `match-case`**

- Để tính được can chi, chúng ta dựa vào quy tắc sau đây:
  - **Can**: Lấy năm sinh chia cho 10 và lấy phần dư.
    - Nếu phần dư bằng 0 tương ứng với Canh, 1 tương ứng với Tân, tiếp tục cho tới 9 tương ứng với Kỷ

  - **Chi**: Lấy năm sinh chia cho 12 và lấy phần dư.
    - Nếu phần dư bằng 0 tương ứng với Thân, 1 tương ứng với Dậu, tiếp tục cho tới 11 tương ứng với Mùi

```python
def get_can(value: int) -> str:

    """
    Trả về Thiên Can dựa trên số dư của năm chia cho 10.

    Parameters:
        value (int): Giá trị year % 10.

    Returns:
        str: Tên Thiên Can tương ứng.
    """
    # Sử dụng match-case để kiểm tra từng giá trị
    match value:
        case 0: return "Canh"
        case 1: return "Tân"
        case 2: return "Nhâm"
        case 3: return "Quý"
        case 4: return "Giáp"
        case 5: return "Ất"
        case 6: return "Bính"
        case 7: return "Đinh"
        case 8: return "Mậu"
        case 9: return "Kỷ"


def get_chi(value: int) -> str:
    """
    Trả về Địa Chi dựa trên số dư của năm chia cho 12.

    Parameters:
        value (int): Giá trị year % 12.

    Returns:
        str: Tên Địa Chi tương ứng.
    """

    # Sử dụng match-case để kiểm tra từng giá trị
    match value:
        case 0: return "Thân"
        case 1: return "Dậu"
        case 2: return "Tuất"
        case 3: return "Hợi"
        case 4: return "Tí"
        case 5: return "Sửu"
        case 6: return "Dần"
        case 7: return "Mão"
        case 8: return "Thìn"
        case 9: return "Tị"
        case 10: return "Ngọ"
        case 11: return "Mùi"


def calculate_can_chi_calendar(year: int) -> str:
    """
    Tính năm Can Chi dựa trên năm dương lịch.

    Parameters:
        year (int): Năm dương lịch cần tính.

    Returns:
        str: Chuỗi Can Chi hoàn chỉnh.
    """

    # Lấy Thiên Can từ số dư chia 10
    can = get_can(year % 10)

     # Lấy Địa Chi từ số dư chia 12
    chi = get_chi(year % 12)

    # Dùng f-string (formatted string literals) để ghép Can và Chi thành kết quả cuối cùng
    return f"{can} {chi}"

# Chạy thử các test case yêu cầu của bài
print("Test case 1:", calculate_can_chi_calendar(2024))
print("Test case 2:", calculate_can_chi_calendar(2023))
print("Test case 3:", calculate_can_chi_calendar(1997))
```

```python
def calculate_can_chi_calendar_optimized(year: int) -> str:
    """
    Tính năm Can Chi từ năm dương lịch bằng cách sử dụng tuple indexing.

    Hàm sử dụng:
    - year % 10 để xác định Thiên Can
    - year % 12 để xác định Địa Chi

    Parameters:
        year (int): Năm dương lịch cần tính Can Chi.

    Returns:
        str: Chuỗi Can Chi hoàn chỉnh.
    """

    # Danh sách Thiên Can theo thứ tự tương ứng với số dư chia 10
    can_list = (
        "Canh", "Tân", "Nhâm", "Quý", "Giáp",
        "Ất", "Bính", "Đinh", "Mậu", "Kỷ"
    )

    # Danh sách Địa Chi theo thứ tự tương ứng với số dư chia 12
    chi_list = (
        "Thân", "Dậu", "Tuất", "Hợi",
        "Tí", "Sửu", "Dần", "Mão",
        "Thìn", "Tị", "Ngọ", "Mùi"
    )

    # Lấy Thiên Can bằng cách truy cập phần tử theo index
    can = can_list[year % 10] # year % 10 sẽ tạo ra giá trị từ 0 -> 9
    chi = chi_list[year % 12] # year % 12 sẽ tạo ra giá trị từ 0 -> 11

    # Dùng f-string (formatted string literals) để ghép Can và Chi thành kết quả cuối cùng
    return f"{can} {chi}"

# Chạy thử các test case để kiểm chứng
print("Test case 1:", calculate_can_chi_calendar_optimized(2024))
print("Test case 2:", calculate_can_chi_calendar_optimized(2023))
print("Test case 3:", calculate_can_chi_calendar_optimized(1997))
```

- Về mặt performance, cách dùng **tuple/list indexing** thường tốt hơn **match-case** cho bài toán kiểu mapping cố định như Can Chi:
  
  1.  **match-case**: Python phải tốn thời gian để thực hiện nhiều comparison -> càng nhiều case → càng nhiều lần check -> Độ phức tạp là `O(n)` -> với `n` là số lượng case
  2. **tuple/list indexing**: Python chỉ cần tính **index** và truy cập trực tiếp phần tử theo index -> Không cần so sánh từng case.  Độ phức tạp là `O(1)`
